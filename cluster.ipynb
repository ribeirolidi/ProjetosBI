```python
def agrupa_portes(df_agrup, df):
    """Ajusta cluster G e M menores que 10 M ou P """

    # Contagem inicial por grupo
    contagem_inicial_por_grupo = df_agrup
    
    # Aplicando as substituições com base nas condições
    grupos_menos_10 = contagem_inicial_por_grupo[(contagem_inicial_por_grupo['PORTE'] == 'G') & (contagem_inicial_por_grupo['Contagem']<10)]
    for _, grupo in grupos_menos_10.iterrows():
        df.loc[
            (df['PORTE'] == 'G') & 
            (df['ESTRUTURA_OPER0'] == (grupo['ESTRUTURA_OPER0'])) & 
            (df['ESTRUTURA_OPER1'] == (grupo['ESTRUTURA_OPER1'])) & 
            (df['COD_ESTRUTURA_MERC2'] == (grupo['COD_ESTRUTURA_MERC2'])), 'PORTE'] = 'M'

    # Contagem após as atualizações
    df_agrup = df.drop(['DATA', 'AVG_ESTOQUE_VLR','AVG_ESTOQUE_QTD'], axis=1)
    df_agrup = df_agrup.drop_duplicates()
    contagem_final_por_grupo = df_agrup.groupby(['ESTRUTURA_OPER0', 'ESTRUTURA_OPER1','ESTRUTURA_OPER2', 'COD_ESTRUTURA_MERC2', 'PORTE']).size().reset_index(name='Contagem')
    

    # Aplicando as substituições com base nas condições
    grupos_menos_10 = contagem_final_por_grupo[(contagem_final_por_grupo['PORTE'] == 'P') & (contagem_final_por_grupo['Contagem']<10)]
    for _, grupo in grupos_menos_10.iterrows():
        df.loc[
            (df['PORTE'] == 'P') & 
            (df['ESTRUTURA_OPER0'] == grupo['ESTRUTURA_OPER0']) & 
            (df['ESTRUTURA_OPER1'] == grupo['ESTRUTURA_OPER1']) & 
            (df['COD_ESTRUTURA_MERC2'] == grupo['COD_ESTRUTURA_MERC2']), 'PORTE'] = 'M'


    # Contagem após as atualizações
    df_agrup = df.drop(['DATA', 'AVG_ESTOQUE_VLR','AVG_ESTOQUE_QTD'], axis=1)
    df_agrup = df_agrup.drop_duplicates()
    contagem_final_por_grupo = df_agrup.groupby(['ESTRUTURA_OPER0', 'ESTRUTURA_OPER1','ESTRUTURA_OPER2', 'COD_ESTRUTURA_MERC2', 'PORTE']).size().reset_index(name='Contagem')
    

    # Aplicando as substituições com base nas condições
    grupos_menos_10 = contagem_final_por_grupo[(contagem_final_por_grupo['PORTE'] == 'M') & (contagem_final_por_grupo['Contagem']<10)]
    for _, grupo in grupos_menos_10.iterrows():
        df.loc[
            (df['PORTE'] == 'M') & 
            (df['ESTRUTURA_OPER0'] == grupo['ESTRUTURA_OPER0']) & 
            (df['ESTRUTURA_OPER1'] == grupo['ESTRUTURA_OPER1']) & 
            (df['COD_ESTRUTURA_MERC2'] == grupo['COD_ESTRUTURA_MERC2']), 'PORTE'] = 'P'

    return df
```
